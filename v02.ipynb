```python
import cv2
import numpy as np 
```

```python
def procesar_imagen(img):
  image=cv2.imread(img,cv2.IMREAD_GRAYSCALE)
  alto,ancho=image.shape
  left_padding=(500-ancho)//2
  right_padding=500-left_padding-ancho
  new_img=cv2.copyMakeBorder(image,0,0,left_padding,right_padding,cv2.BORDER_CONSTANT,value=[255,255,255])
  return new_img
```

Para el procesamiento de imagenes, la idea basica es:
- Lo paso a blanco y negro
- Divido numero de casillas con negro sobre el total de casillas de la imagen

## La funcion de negros (nombre en proceso)
Aca declaro y uso la funcion de proporciones. La matriz ahora solo es blanco y negro

```python
def conteo(img):
  negros=np.sum(img==0)
  alto,ancho=img.shape
  tamanyo=alto*ancho
  return (negros/tamanyo*100)
```
