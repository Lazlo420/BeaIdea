```python
import cv2
from PIL import Image
def contar_pixeles_negros(imagen):
    # Cargar la imagen en blanco y negro
    img = cv2.imread(imagen, cv2.IMREAD_GRAYSCALE)

    # Obtener la altura y ancho de la imagen
    alto, ancho = img.shape

    # Contar píxeles negros
    total_pixeles_negros = cv2.countNonZero(img)

    # Calcular el porcentaje de píxeles negros
    porcentaje_negro = (total_pixeles_negros / (alto * ancho)) * 100

    return porcentaje_negro
```

```python
import cv2
from PIL import Image

def count_percentage_black(image):
    # Convert the image to grayscale
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Get the height and width of the image
    alto, ancho = img_gray.shape

    # Count black pixels (pixels with values less than 200)
    total_black_pixels = (alto * ancho) - cv2.countNonZero(cv2.threshold(img_gray, 200, 255, cv2.THRESH_BINARY)[1])

    # Calculate the percentage of black pixels
    percentage_black = (total_black_pixels / (alto * ancho)) * 100

    return percentage_black, img_gray, alto, ancho

# Load the image
i=1
image_path = "Anuncios/test"+str(i)+".png"
img = cv2.imread(image_path)

# Call the function to get percentage, grayscale image, height, and width
percentage_black, img_gray, alto, ancho = count_percentage_black(img)
print(f"TEST #{i}")
# Print the percentage of black pixels
print("Percentage of black pixels: {:.2f}%".format(percentage_black))

print("Image dimensions - Height: {}, Width: {}".format(alto, ancho))
i=i+1
# Display the original image
image = Image.open(image_path)
image
```
